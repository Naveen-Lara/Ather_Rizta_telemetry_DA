# ***Correlation heatmap***

```python
cols=df.select_dtypes(include="float64").columns.tolist()
cor_matrix=df[cols].corr()
plt.figure(figsize=(12,4))
sns.heatmap(cor_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

### *Motor_Current and Battery_Voltage are negatively correlated*
        - High motor current tends to reduce battery voltage, indicating voltage sag under load.
### *Motor_Current and Rear_PSI are negatively correlated*
        - Suggests that low rear tyre pressure may be associated with higher motor load, possibly due to increased rolling resistance.
### *Discharge_Current and Battery_Voltage are negatively correlated*
        - As discharge current increases, battery voltage drops — expected behavior under heavy load.
### *Battery_Temp and Motor_Temp are strongly positively correlated*
        - Both rise together, likely due to simultaneous strain on motor and battery during high current demand.
### *Estimated_SOC and Battery_Voltage are strongly positively correlated*
        - Confirms that battery voltage is a reliable proxy for SOC in your estimation model.
### *Efficiency and Estimated_Range_km are positively correlated*
         - Confirms that higher efficiency leads to longer range, validating your estimation formula.
### *Motor_Current and Efficiency are positively correlated (unexpected)*
        - Might indicate that high-efficiency states still require high motor current — potentially due to aggressive riding or terrain

## **-->** *Motor_current and Battery_voltage are negatively correlated*
## **-->** *Motor_current and Rear_Psi are negatively correlated*
## **-->** *Discharge_current and Battery_voltage are negatively correlated*
## **-->** *Soc and Rear_psi Have moderate positive correlation 0.4 that is 40%*

```python
corr_matrix = df.corr(numeric_only=True)
corr_pairs = corr_matrix.unstack()
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
corr_pairs
```

```python
positive_corr = corr_pairs[corr_pairs > 0.5].sort_values(ascending=False).drop_duplicates()
print("Top Positive Correlations:\n", positive_corr)
```

```python
negative_corr = corr_pairs[corr_pairs < -0.5].sort_values()
print("Top Negative Correlations:\n", negative_corr)
```

# *Does tyre pressure impact SOC consumption over longer rides?*

```python
soc_per_ride = df.groupby('Ride_id').agg({
    'Front_PSI': 'mean',
    'Rear_PSI': 'mean',
    'Estimated_SOC': ['first', 'last']
})

soc_per_ride.columns = ['Front_PSI', 'Rear_PSI', 'SOC_Start', 'SOC_End']
soc_per_ride['SOC_Drop'] = soc_per_ride['SOC_Start'] - soc_per_ride['SOC_End']

# Correlation check
print(soc_per_ride[['Rear_PSI', 'SOC_Drop']].corr())
```

## *The rear_psi and Soc drop is negatively correlated so lower the tyre pressure higher the SOC drop*

# *Soc degradation*

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

sns.regplot(x='Motor_Current', y='Estimated_SOC', data=df, ax=axes[0], line_kws={'color': 'red'})
axes[0].set_title('Motor_Current vs SOC')

sns.regplot(x='Battery_Temp', y='Estimated_SOC', data=df, ax=axes[1], line_kws={'color': 'blue'})
axes[1].set_title('Battery_Temp vs SOC')

plt.tight_layout()
plt.show()
```

### *Is Smart Eco mode significantly more efficient in terms of Wh/km or SOC drop?*

```python
# Per-ride SOC drop
soc_stats = df.groupby(['Ride_id', 'Ride_Mode'])['Estimated_SOC'].agg(['first','last'])
soc_stats['SOC_Drop'] = soc_stats['first'] - soc_stats['last']
soc_by_mode = soc_stats.groupby('Ride_Mode')['SOC_Drop'].mean()
print(soc_by_mode)
```

#### *Obvisouly the efficiency is more in Smart ECO than in Zip mode*
